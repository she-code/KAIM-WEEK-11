3. Clean and Prepare Data

```python
data = data.iloc[2:].reset_index(drop=True)

# Step 1: Set the 'Ticker' column as the index (contains dates)
data.set_index('Ticker', inplace=True)

# Step 2: Select only the 'Close' columns for each asset
close_columns = ['TSLA.3', 'BND.3', 'SPY.3']  # Correspond to 'Close' for TSLA, BND, SPY
prices = data[close_columns]

# Step 3: Rename columns to single-level asset names
prices.columns = ['TSLA', 'BND', 'SPY']

# Step 4: Convert data to numeric, handling any non-numeric values
prices = prices.apply(pd.to_numeric, errors='coerce')

# Step 5: Drop any rows with NaN values (if necessary)
prices = prices.dropna()
prices
```
